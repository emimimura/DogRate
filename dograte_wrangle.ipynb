### API for Additional Tweets Data

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)
```

```python
# Initialize empty list to store tweets: tweets_data
consumer_key = 
consumer_secret = 
access_token = 
access_secret = 

tweets_data = []
deleted = []

for tweet_id in tweet_ids:
    print(tweet_id)
    try:
        tweet = api.get_status(tweet_id)
        tweets_data.append(tweet)
    except:
        deleted.append(tweet)
```

```python
# Create empty list for each twitter status to be stored
json_strs = []

# Appending each string to the text file
for tweet_data in tweets_data:
    json_str = json.dumps(tweet_data._json)
    json_strs.append(json_str)
```

```python
# Writing text file from JSON
with open('tweet_json.txt', 'w') as outfile:
    json.dump(json_strs, outfile)
```

```python
# Creating an object from text file
with open('tweet_json.txt', 'r') as fp:
    obj = json.load(fp)
```

```python
# From the object of list to make a dictionary
tweets_dict = []
for line in obj:
    tweet_line = json.loads(line)
    tweets_dict.append(tweet_line)
```

```python
# Create a pandas DataFrame using the keys from the first line
columns = ['created_at', 'id', 'id_str', 'text', 'truncated', 
           'entities', 'extended_entities', 'source', 
           'in_reply_to_status_id', 'in_reply_to_status_id_str', 
           'in_reply_to_user_id', 'in_reply_to_user_id_str', 
           'in_reply_to_screen_name', 'user', 'geo', 'coordinates', 
           'place', 'contributors', 'is_quote_status', 'retweet_count', 
           'favorite_count', 'favorited', 'retweeted', 
           'possibly_sensitive', 'possibly_sensitive_appealable', 'lang']

df_tweets = pd.DataFrame(tweets_dict, columns=columns)
```

```python
# Correcting the rows with two stages when there should be only one
master.loc[master.tweet_id == 854010172552949760, 'floofer'] = 'None'
master.loc[master.tweet_id == 817777686764523521, 'pupper'] = 'None'
master.loc[master.tweet_id == 801115127852503040, 'doggo'] = 'None'
master.loc[master.tweet_id == 785639753186217984, 'pupper'] = 'None'
master.loc[master.tweet_id == 751583847268179968, 'pupper'] = 'None'
master.loc[master.tweet_id == 855851453814013952, 'doggo'] = 'None'

# Changing the two-dog rows to "doggo_pupper" and replacing "pupper" with "None".
doggo_pupper_array = [808106460588765185, 759793422261743616, 741067306818797568, 733109485275860992]
def doggo_pupper(id):
    master.loc[master.tweet_id == id, 'doggo'] = 'doggo_pupper'
    master.loc[master.tweet_id == id, 'pupper'] = 'None'

for id in doggo_pupper_array:
    doggo_pupper(id)
```

```python
# Reference: https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column
conditions = [
    (master['doggo'] == 'doggo'),
    (master['doggo'] == 'doggo_pupper'),
    (master['floofer'] == 'floofer'),
    (master['pupper'] == 'pupper'),
    (master['puppo'] == 'puppo')]
choices = ['doggo', 'doggo_pupper', 'floofer', 'pupper', 'puppo']
master['stage'] = np.select(conditions, choices, default='unknown')

# Dropping the individual stage columns
master = master.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)
```
